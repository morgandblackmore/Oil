I chose to run an analysis on how oil affects inflation, using the consumer price index, and how or if anything the fed rate does affects the price of oil.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
import hvplot.pandas
```

```python
df = pd.read_csv("oil_cpi_fedrate.csv")
```

```python
# convert date column into date format
df['DATE'] = pd.to_datetime(df['DATE'])
```

The first time period I chose to analysis is oil embargo of the 70s into the inflationary period of the 80s to see if there is any type of correlation. I chose this time period because I feel as though it is similar to the geopolitical situation we are currently in.  

```python
# filter rows on the basis of date and locate rows by slicing the DataFrame  
oil_embargo = df[(df['DATE'] >= '3/1/1967') & (df['DATE'] <= '9/1/1986')]
oil_em_corr = oil_embargo.corr()
```

The second time period I chose to analyze is the recession caused by the housing market collapse. I started in 2007 slightly before the collapse and went to 2012 which is when the recovery began. I chose this time period to see if there are any similarities to the present.

```python
real_estate_collapes = df[(df['DATE'] >= '1/1/2007') & (df['DATE'] <= '9/1/2012')]
real_es_corr = real_estate_collapes.corr()
```

The last period I chose to analyze is the current time period begaining before the pandemic to as close to the current date I could get with the available data. 

```python
pandemic = df[(df['DATE'] >= '1/1/2018') & (df['DATE'] <= '2/1/2022')]
pand_corr = pandemic.corr()
```

I used heat maps to visualize the corralations between the 3 variables. 

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 4)) 
sns.heatmap(ax=axes[0], data=oil_em_corr, annot = True).set(title = "Oil Embargo Corrolation")
sns.heatmap(ax=axes[1], data=real_es_corr, annot = True).set(title = "Real Estate Collapses Corr")
sns.heatmap(ax=axes[2], data=pand_corr, annot = True).set(title = "Pandemic Corr") 
plt.show()
```

Its interesting that in the first time period the fed rate correlates more with the price of oil than the consumer price index does. And in the second time period the fed rate and CPI have almost identical correlation to oil but not to each other. And in the current situation, the correlation between oil and the CPI has a closer correlation than the federate.

```python
oil_embargo.hvplot(
    x='DATE',
    title='Oil Embargo')
```

Despite our heat map showing some correlation between the fedrate and oil the plot doesn't show much correlation.

```python
real_estate_collapes.hvplot(
    x='DATE',
    title='Real Estate Collapse')
```

During the recession of 2008 the price of oil and inflation move in a similar pattern when you compare peaks and valleys. 

```python
pandemic.hvplot(
    x='DATE',
    title='Pandemic')
```

There is some corralation based on the chart. It makes sense that inflation moves with the price of oil. As oil has a direct affect on the cost of moving goods and delivering services. And fuel prices are included in the consumer price index.
